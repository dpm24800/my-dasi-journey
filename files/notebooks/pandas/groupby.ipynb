```python
import pandas as pd

data = {
    'Region': ['East', 'West', 'East', 'North', 'West', 'East'],
    'Sales': [200, 120, 340, 300, 250, 400],
    'Profit': [20, 15, 50, 30, 25, 60]
}

df = pd.DataFrame(data)
df
```

```python
grouped = df.groupby('Region')['Sales'].sum()
grouped
```

```python
data = {
    'Region': ['East', 'East', 'West', 'West', 'North', 'North'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Sales': [200, 340, 120, 250, 300, 150]
}

df = pd.DataFrame(data)
df
```

```python
df.groupby(['Region']).sum() 
```

```python
df.groupby(['Region', 'Product']).sum() # Grouping by Multiple Columns
```

```python
df.groupby(['Region', 'Sales']).sum() # Grouping by Multiple Columns
```
