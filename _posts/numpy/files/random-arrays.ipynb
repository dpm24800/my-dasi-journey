# Random Arrays

#### x. `np.random.rand()`

```python
import numpy as np

arr17 = np.random.rand(5) # between 0-1 (uniform distribution) 
print(arr17)
```

```python
arr18 = np.random.rand(5, 5) # between 0-1
print(arr18)
```

```python
arr19 = np.random.rand(2, 5, 5) # between 0-1
print(arr19)
```

#### x. `np.random.randint()`

```python
arr20 = np.random.randint(1, 10, size=10, dtype=np.uint8) # lowest (inclusive), highest (exclusive)
print(arr20)
```

```python
arr21 = np.random.randn(10) # normal distribution (most numbers are close to 0, fewer)
print(arr21)
```

```python
np.random.seed(42)
print(np.random.rand(3))
```
