### Initial Placeholders

#### `np.zeros()`

```python
import numpy as np

arr8 = np.zeros(5)

print(arr8)
print(arr8.dtype)
```

```python
arr8 = np.zeros(5, dtype='int')

print(arr8)
```

```python
arr9 = np.zeros((3, 4))
# arr9.shape
# arr9.ndim
print(arr9)
```

```python
arr9 = np.zeros((3, 4, 4),dtype=int)
print(arr9)
```

#### `np.ones()`

```python
arr10 = np.ones((3, 4, 4), dtype=int)
print(arr10)
```

#### `np.eye()` /identity matrix

```python
arr11 = np.eye((4), dtype=int)
print(arr11)
```

```python
arr12 = np.eye(3,5)
print(arr12)
```
