```python
import numpy as np
import pandas as pd


data = {
    'f_name': ['Dipak', 'Ramesh', 'Sita', 'Gita', 'Anup', 'Gita', 'Mira', 'Sundar', 'Manoj', 'Ganesh'],
    'l_name': ['Prasai', 'Ghale', 'Mahatara', np.nan, 'Bagale', 'Tamang', 'Shah', 'Gurung', 'Ghale', np.nan],
    'age': [32, 45, 21, 15, 36, np.nan, 85, 41, 32, 12],
    'weight': [65, 72, 55, 40, 58, 62, np.nan, 65, 55, 20], 
    'balance': [150000, 200000, 25000, 5000, 500000, 200000, 50000, 256000, 78000, 25000]
}

df = pd.DataFrame(data)
```

```python
# Add new column
df['full_name'] = df['f_name'] + " " + df['l_name'] # deep
df
```

```python
df0 = df.assign(bonus = df['age'] * 5/10, allowance = df['age'] * 2) # shallow
df0
```
