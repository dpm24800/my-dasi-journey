```python
import numpy as np
import pandas as pd


data = {
    'f_name': ['Dipak', 'Ramesh', 'Sita', 'Gita', 'Anup', 'Gita', 'Mira', 'Sundar', 'Manoj', 'Ganesh'],
    'l_name': ['Prasai', 'Ghale', 'Mahatara', np.nan, 'Bagale', 'Tamang', 'Shah', 'Gurung', 'Ghale', np.nan],
    'age': [32, 45, 21, 15, 36, np.nan, 85, 41, 32, 12],
    'weight': [65, 72, 55, 40, 58, 62, np.nan, 65, 55, 20], 
    'balance': [150000, 200000, 25000, 5000, 500000, 200000, 50000, 256000, 78000, 25000]
}

df = pd.DataFrame(data)
```

```python
# Detecting duplicated values
df['balance'].duplicated() # second (or later) occurance is considered as duplicated
```

```python
# Count duplicated values
df['balance'].duplicated().sum()
```

```python
# Drop duplicated values
df['balance'] = df['balance'].drop_duplicates()
df['balance']
```

---

```python
# Count missing/empty values
df['balance'].isna().sum()
```

```python
# Fill null value by zero
df['balance'].fillna(0)
# df.fillna(0)
```

```python
# Fill null value by the mean of the balance
df['balance'].fillna(df['balance'].mean())
```
