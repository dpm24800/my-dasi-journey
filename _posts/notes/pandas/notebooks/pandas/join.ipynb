## Join

```python
import pandas as pd

# Left DataFrame
df1 = pd.DataFrame({
    'emp_id': [1, 2, 3, 4],
    'name': ['Dipak', 'Sita', 'Ram', 'Gita'],
    'dept_id': [101, 102, 101, 103]
})

df1
```

```python
# Right DataFrame
df2 = pd.DataFrame({
    'dept_id': [101, 102, 103],
    'department': ['IT', 'HR', 'Finance']
})

df2
```

```python
# Set index to dept_id for both DataFrames
df1.set_index('dept_id', inplace=True)
df2.set_index('dept_id', inplace=True)

# Join using index
joined_df = df1.join(df2)
joined_df
```
