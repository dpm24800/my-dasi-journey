## merge

```python
import pandas as pd

# Left DataFrame
df1 = pd.DataFrame({
    'emp_id': [1, 2, 3, 4],
    'name': ['Dipak', 'Sita', 'Ram', 'Gita'],
    'dept_id': [101, 102, 101, 103]
})

df1
```

```python
# Right DataFrame
df2 = pd.DataFrame({
    'dept_id': [101, 102, 103],
    'department': ['IT', 'HR', 'Finance']
})

df2
```

```python
# Merge based on common column 'dept_id'
merged_df = pd.merge(df1, df2, on='dept_id', how='inner')

merged_df
```

```python
merged_df = pd.merge(df1, df2, on='dept_id', how='outer')

merged_df
```

```python
merged_df = pd.merge(df1, df2, on='dept_id', how='left')

merged_df
```

```python
merged_df = pd.merge(df1, df2, on='dept_id', how='right')

merged_df
```
