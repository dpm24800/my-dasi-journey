```python
import pandas as pd

df1 = pd.DataFrame({'name': ['sachin', 'virat', 'sourav'], 
                    'wickets': [98, 97, 96], 
                    'year': [1990, 2000, 2010]})

df2 = pd.DataFrame({'name': ['mark', 'robert', 'adam'], 
                    'wickets': [90, 95, 92], 
                    'year': [1990, 2000, 2010]})

df3 = pd.concat([df1, df2], axis=0)
df4 = pd.concat([df1, df2], axis=1)

print("df1:\n", df1, "\n")
print("df2:\n", df2, "\n")

print("pd.concat([df1, df2], axis=0):")
print(df3)

print()
print("pd.concat([df1, df2], axis=1):")
print(df4)
```

```python
import numpy as np
import pandas as pd

data = {
    'id': [1, 4, 5, 3, 2, 4, 6],
    'name': ['emiley', 'bibek', 'chandra', 'amar', 'dipika', 'anmol', 'farhan'] 
}

df = pd.DataFrame(data)
```

```python
df1 = df.sort_values('id')
df2 = df.sort_values('name')
df3 = df.sort_values(by=['id', 'name'], ascending=[True, True])
df4 = df.sort_values(by=['name', 'id'], ascending=[True, True])

print("Original DataSet:"); print(df); print()
print("In ascending order according to id:"); print(df1); print()
print("In descending order according to name:"); print(df2); print()
print("In descending order according to both (id and name):"); print(df3); print()
print("In descending order according to both (name and id):"); print(df4); print()
```
